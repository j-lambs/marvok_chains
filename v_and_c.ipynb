# Building a Markov Chain with Macbeth

## Processing characters

```python
# Read text file
with open('macbeth.txt', 'r', encoding='utf-8') as f:
    text = f.read()

stripped_text = ''.join(char.lower() for char in text if char.isalpha()) # strip text (alphabet characters only and lowercase all)

print(f'First 100 characters:\n{stripped_text[:100]}')
```

## Building the vowel/consonant chain

```python
def is_vowel_or_consonant(char: str):
    if (char == 'a' or char == 'e' or
        char == 'i' or char == 'o' or char == 'u'):
        return('v')
    else:
        return('c')

def get_state_jump(char_1: str, char_2: str): 
    state = list()
    if is_vowel_or_consonant(char_1) == 'v':
        state.append(0)
    else:
        state.append(1)
    if is_vowel_or_consonant(char_2) == 'v':
        state.append(0)
    else:
        state.append(1)
    return(state)

def sum_jumps(matrix):
    """
    Given a 2-D array of numbers, sum the entries
    """
    sum = 0
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[0])):
            sum += matrix[i][j]
    return(sum)

# build 2x2 matrix with total number of 'jumps' 
total = [[0, 0], [0, 0]]

for i in range(1, len(stripped_text)):
    cur_state = get_state_jump(stripped_text[i-1], stripped_text[i])
    total[cur_state[0]][cur_state[1]] += 1
print(total)

total_jumps = sum_jumps(total)
print(total_jumps)

# turn into markov matrix (with probabilities of jumps)
for i in range(0, len(total)):
        for j in range(0, len(total[0])):
            total[i][j] /= total_jumps
            total[i][j] = round(total[i][j], 4)
print(total)

```
